```octave
% gda00_07
% example of vector and matrix algebra

% define matrices and vectors
a=[1, 3, 5]';
b=[2, 4, 6]';
M=[ [1, 0, 2]', [0, 1, 0]', [2, 0, 1]'];
N=[ [1, 0, -1]', [0, 2, 0]', [-1, 0, 3]'];
I = eye(3);
I

% addition and subtraction
S = M+N;
S
D = M-N;
D

% dot product
s = a'*b;
s

% outer product
T = a*b';
T

% matrix times vector
c = M*a;
c

% matrix times matrix
P = M*N;
P

% element-wise multiplication of two vectors
d = a.*b;
d
```

```octave
% gda00_09
% example of the matrix inverse, determinant, eigenvalues

% define a matrix and a vector
A = [ [1, 5, 13]', [2, 7, 17]', [3, 11, 19]'];
b = [1, 2, 3]';

A

% compute inverse
B = inv(A);
B

% check result
A*B
B*A

% determinant
d = det(A);
d

% solve b = A*c
c = A\b;
c

% check result
error=b-A*c;
error

% define matrix B
B = [ [1, 3, 4]', [2, 3, 2]', [0, 0, 4]'];
B

% D=B*inv(A)
D=B/A;
D

% define symmetric matrix M
M = [ [1, 2, 0]', [2, 2, 0]', [0, 0, 4]'];
M

% eigenvalues and eigenvectors
[V,LAMBDA] = eig(M);
V
LAMBDA

% check orthonormality
V'*V
```

```octave
% gda00_10
% example of a for loop

% define a matrix M and a vector a
M = [ [1, 4, 7]', [2, 5, 8]', [3, 6, 9]'];
a = [0, 0, 0]';

% use a for loop to extract the diagonal elements of
% the matrix M into the column-vector a
for i = [1:3]
    a(i) = M(i,i);
end


M
a
```

```octave
% gda00_11
% example of two nested for loops

% define a matrices M and N
M = [ [1, 4, 7]', [2, 5, 8]', [3, 6, 9]'];
N = [ [0, 0, 0]', [0, 0, 0]', [0, 0, 0]'];

% use two nested for loop to reverse the order
% of the elements in each row of the matrix M
for i = [1:3]
for j = [1:3]
    N(i,4-j) = M(i,j);
end
end

M
N
```

```octave
% gda00_12
% example of a for loop containing a consitional statement

% define vectors a and b
a = [ 1, 2, 1, 4, 3, 2, 6, 4, 9, 2, 1, 4 ]';

% use two nested for loop to reverse the order
% of the elements in each row of the matrix M
b = a;
for i = [1:12]
    if ( a(i) >= 6 )
        b(i) = 6;
    else
        b(i) = a(i);
    end
end

% display vectors side by side
[a, b]
```

```octave
% gda00_13
% avoiding loops

% define a matrix M
M = [ [1, 4, 7]', [2, 5, 8]', [3, 6, 9]'];

% extract diagonal
a = diag(M);
M
a

% reverse order of rows
N = fliplr(M);
M
N

% another way of reversing order order of rows
N = M(:,end:-1:1);
M
N

% define vector a
a = [ 1, 2, 1, 4, 3, 2, 6, 4, 9, 2, 1, 4 ]';

% clip to 6
b=a;
b(find(a>6))=6;
[a, b]

% alternative way to clip to 6
b = a.*(a<6)+6.*(a>=6);
[a, b]
```

```octave
% gda00_14
% 
% load and plot 1965-2010 global temperature data

% Hansen, J., Mki. Sato, R. Ruedy, K. Lo, D.W. Lea, and M. Medina-Elizade,
% 2006: Global temperature change. Proc. Natl. Acad. Sci., 103, 14288-14293,
% doi:10.1073/pnas.0606291103. 

% load
D=load('../data/global_temp.txt');
t=D(:,1);
d=D(:,2);
N=length(d);

disp(sprintf('Number of data: %d\n', N));

% display first few lines
disp('first few lines of data');
for i=[1:4]
    disp(sprintf('%f  %f',t(i),d(i)));
end

% plot data
figure(1);
clf;
set(gca,'LineWidth',3);
hold on;
axis( [1965, 2010, -0.5, 1.0] );
plot(t,d,'r-','LineWidth',3);
plot(t,d,'ko','LineWidth',3);
xlabel('calendar year');
ylabel('temperature anomaly, deg C');
title(sprintf('temperature data from %d to %d',t(1),t(N)))
```
